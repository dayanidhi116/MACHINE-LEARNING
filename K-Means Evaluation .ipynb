```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
import matplotlib.pyplot as plt

# Create sample data
X, _ = make_blobs(n_samples=500, centers=4, cluster_std=1.0, random_state=42)

# Apply K-Means
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_

# Evaluation Metrics
inertia = kmeans.inertia_  # Sum of squared distances to nearest cluster center
silhouette = silhouette_score(X, labels)
calinski = calinski_harabasz_score(X, labels)
davies = davies_bouldin_score(X, labels)

# Print results
print(f"ðŸŒ€ Inertia (SSE): {inertia:.2f}")
print(f"ðŸ“ˆ Silhouette Score: {silhouette:.2f}")
print(f"ðŸ“Š Calinski-Harabasz Index: {calinski:.2f}")
print(f"ðŸ“‰ Davies-Bouldin Index: {davies:.2f}")

# Plot clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=30)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c='red', s=200, marker='X', label='Centroids')
plt.title("K-Means Clustering Visualization")
plt.legend()
plt.grid(True)
plt.show()
```
